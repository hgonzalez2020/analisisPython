```python
ssma = pd.read_csv("/content/drive/MyDrive/bases de datos/Students Social Media Addiction.csv")  #lee el archivo que esta en drive
ssma
```

```python
ssma_1=ssma [['Age','Gender','Avg_Daily_Usage_Hours','Most_Used_Platform']]  #filtra la tabla con 4 columnas
ssma_1
```

```python
avg = ssma_1[ssma_1['Avg_Daily_Usage_Hours']>4.0]   #filtra unicamante a quies usan mas de 4 horas diarias las redes
avg
```

```python
ssma_1 = ssma_1.rename(columns={'Avg_Daily_Usage_Hours':'Horas', "Most_Used_Platform":'Plataforma',"Age":"Edad","Gender":"Sexo"})  #renomba los nombres de los campos
ssma_1
```

```python
grouppssm_2 = ssma_1.groupby(by="Plataforma")['Horas'].max()  #maximo de horas que ultizan las plataformas
grouppssm_2
```

```python
grouppssm_2 = ssma_1.groupby(by="Plataforma")['Horas'].mean() #saca el promedio de horas por plataforma
grouppssm_2
```

```python
grouppssm_2 = ssma_1.groupby(by="Plataforma")['Horas'].count() #saca el numero de personas que utilizan las redes sociales
grouppssm_2
```

```python
sssma_3 = ssma_1.groupby("Plataforma")["Horas"].size()  #filtra las horas por cada plataformas
ssma_4 =sssma_3.sort_values(ascending=True)  #ordena los valores de manera ascendente
ssma_4
```
